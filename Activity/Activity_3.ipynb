```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('C:/Users/Administrator/Downloads/platform_data.csv')

df.columns = [
    "Platform Name", "Courses", "Free/Paid", "Duration", 
    "Marketing - Cost per lead", "Marketing - Conversion Rate", "Marketing - Student Enrolled", 
    "Marketing - Marketing ROI", "Services - Course Completion Rate", 
    "Services - Engagement Metrics", "Services - Placement Rate", 
    "Services - Course Variety", "Branding - Brand Reach", 
    "Branding - Share of Voice", "Branding - Branded Search Trends", 
    "Branding - Referral Rate", "Customer Expectation - Customer Satisfaction Score", 
    "Customer Expectation - Customer Effort Score", 
    "Customer Expectation - Net Promoter Score"
]
df.columns = df.columns.str.strip()
```

```python
df.head(3)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Platform Name'], df['Services - Course Completion Rate'], color='red')
plt.xlabel('Platform Name')
plt.ylabel('Course Completion Rate')
plt.title('Course Completion Rate per Platform')
plt.xticks(rotation=0)  
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.barh(df['Platform Name'], df['Services - Placement Rate'], color='green')
plt.xlabel('Platform Name')
plt.ylabel('Placement Rate')
plt.title('Placement Rate per Platform')
plt.xticks(rotation=90)  
plt.tight_layout()
plt.grid()
plt.show()
```
