```python
global_dict = {}

def deposit(key=0, value=0):
    if not isinstance(key,int):
        return "Please enter a valid key"

    if not isinstance(value,(int,float)) :
        return "Can only deposit numerical values"

    if value<0:
        return "Can only deposit positive integers"

    if key in global_dict:
        global_dict[key] += value

    if key<=0 or value<=0:
        return "Please enter a valid key and value"

    else:
        global_dict[key] = value
    return f"Total balance: {global_dict[key]}"

def withdraw(key=0, value=0):
    if not isinstance(value,int) :
        return "Can only withdraw numerical values"

    if value<0:
        return "Can only withdraw positive integers"

    if key<=0 or value<=0:
        return "Please enter a valid key and value"

    if key in global_dict:
        if global_dict[key]>=value:
            global_dict[key]-=value
            return f"Withdraw amount {value}"

        else:
            return "Not enough balance available"
        return f"Remaining balance: {global_dict[key]}"
    else:
        return "User does not exist"

def check_balance(key=0):
    if not isinstance(key,int):
        return "Please enter a valid key"
    if key in global_dict:
        return f"Remaining balance: {global_dict[key]}"
    if key<=0:
        return "Please enter a valid key"
    else:
        return "User does not exist"





import unittest
class TestAddFunction(unittest.TestCase):
  def test_deposite(self):
    self.assertEqual(deposit(1,1000),"Total balance: 1000")
    self.assertEqual(deposit(1,"abc"),"Can only deposit numerical values")
    self.assertEqual(deposit(1,-1000),"Can only deposit positive integers")
    print("All deposit test cases passed ")

  def test_withdraw(self):
    deposit(2,1000)
    self.assertEqual(withdraw(2,500),"Withdraw amount 500")
    self.assertEqual(withdraw(2,'abc'),"Can only withdraw numerical values")
    self.assertEqual(withdraw(2,1000),"Not enough balance available")
    self.assertEqual(withdraw(2,-1000) ,"Can only withdraw positive integers")
    print("All withdraw test cases passed")


  def test_check_balance(self):
    deposit(3,1000)
    self.assertEqual(check_balance(3),"Remaining balance: 1000")
    self.assertEqual(check_balance(100),"User does not exist")
    print("Check balance test cases passed")

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)

```
