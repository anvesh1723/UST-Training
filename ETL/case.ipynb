```python
#15. Create Risk Score (Custom Analysis) 
```

```python

df['risk_score'] = (df['chol'] / 200) + (df['trestbps'] / 120) + df['oldpeak']

conditions = [
    (df['risk_score'] < 2),
    (df['risk_score'] >= 2) & (df['risk_score'] < 4),
    (df['risk_score'] >= 4)
]
choices = ['Low Risk', 'Medium Risk', 'High Risk']

df['risk_category'] = np.select(conditions, choices)


df['risk_category'].value_counts().plot(kind='bar', title="Risk Score Distribution")# can add colors
plt.ylabel('Count')
plt.show()
```
