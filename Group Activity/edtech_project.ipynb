```python
9. Course Success Prediction Model
```

```python
#convert learning model to number
df["mode_num"] = df["mode"].map({
    "Online": 3,
    "Hybrid": 2,
    "Offline": 1
})
```

```python
df["enroll_norm"] = df["enrollemnt count (past 3)"] / 100
```

```python
#create success score
df["success_score"] = (
    0.4 * df["rating"]
    + 0.25 * df["enroll_norm"]
    + 0.2 * df["placement_num"]
    + 0.1 * df["cert_num"]
    + 0.05 * df["mode_num"]
)
```

```python
#classify success rate
def success_level(x):
    if x >= 4:
        return "High Success"
    elif x >= 3:
        return "Moderate"
    else:
        return "Low"
df["success_level"] = df["success_score"].apply(success_level)
```

```python
#sort the values
success_df = df.sort_values(
    by="success_score",
    ascending=False
)
success_df
```

```python
#show result ontoit
success_df[[
    "company",
    "course",
    "rating",
    "enrollemnt count (past 3)",
    "placement assistance(past 3)",
    "certification",
    "mode",
    "success_score",
    "success_level"
]]
```

```python
best = success_df.iloc[0]
print("Best Success Course")
print(best)
```

```python
10. Smart Institute Ranking System
```

```python
#normalize to enrollment to doo check don
df["enroll_norm2"] = df["enrollemnt count (past 3)"] / df["enrollemnt count (past 3)"].max()
```

```python
#normalize cost afferdly to done to itl
df["cost_norm2"] = 1 - (df["cost (INR)"] / df["cost (INR)"].max())
```

```python
df["inst_score"] = (
    0.35 * df["rating"]
    + 0.25 * df["placement_num"]
    + 0.20 * df["enroll_norm2"]
    + 0.10 * df["cert_num"]
    + 0.10 * df["cost_norm2"]
)
```

```python
rank_df = df.sort_values(
    by="inst_score",
    ascending=False
)
rank_df
```

```python
#create category don to itltu
def category(x):
    if x >= 3.5:
        return "Premium Institute"
    elif x >= 2.5:
        return "Best Value"
    else:
        return "Emerging"

df["category"] = df["inst_score"].apply(category)
```

```python
#show final rank
rank_df = df.sort_values(
    by="inst_score",
    ascending=False
)
rank_df[[
    "company",
    "course",
    "rating",
    "placement assistance(past 3)",
    "certification",
    "cost (INR)",
    "inst_score",
    "category"
]]
```

```python
#top instits choose to do itl
best = rank_df.iloc[0]
print("Top Institute")
print(best)
```
