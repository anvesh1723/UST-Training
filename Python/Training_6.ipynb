```python
import matplotlib.pyplot as plt
import numpy as np

data=np.random.random_sample(5)

#list1=[1,2,3,-1,-2,1,1,2,3,1,1,0,0,5,6,7,-2,-3]

plt.hist(data,color='green',edgecolor='yellow',bins=5)
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('Histogram Plot')

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,15,13,17,20]

plt.scatter(x,y,marker='o',color='blue') # represent relationship, and correlations

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot')
plt.show()
```

```python
#pie chart

categories=['Transport','housing','food','utilites','entertainment']
expenses=[1500,1000,500,300,200]

plt.figure(figsize=(5,16))

plt.pie(expenses,labels=categories,autopct='%2.1f%%')

plt.title('expenses')
#plt.savefig()
plt.legend()
plt.show()
```

```python
# box plot
import matplotlib.pyplot as plt

data=[12,14,15,19,20,22,27,35,45,100,-2]

plt.boxplot(data)

plt.title('Bloxplot example with sample data')
plt.xlabel('data')
plt.ylabel('show')
plt.show()
```

```python
import matplotlib.pyplot as plt

quarters=['Q1','Q2','Q3','Q4']
electronics=[10000,12000,11000,10500]
clothing=[5000,6000,7500,8500]
appliances=[3000,6000,8000,9000]

plt.figure(figsize=(10,6))
plt.stackplot(quarters,electronics,clothing,appliances,labels=['electronics','clothing','appliances'])

plt.xlabel('quarters')
plt.ylabel('sales')
plt.title('product category sales over quarters')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#generate data
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2+y**2))

#3d scatter plot
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c=z,cmap='viridis')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(-x)
#create a 2x2 subplot
fig,axs=plt.subplots(2,2, figsize=(10,8))

#plot data on each subplot

axs[0,0].plot(x,y1,'r') #top-left subplot
axs[0,0].set_title('Sine Wave')
axs[0,1].plot(x,y2,'b') #top-right
axs[0,1].set_title('cosine wave')
axs[1,0].plot(x,y3,'g') # bottom-left
axs[1,0].set_title('Tangent wave')
axs[1,1].plot(x,y4,'m')
axs[1,1].set_title('Exponential Decay')

plt.tight_layout() #adjust layout
plt.show()
```
