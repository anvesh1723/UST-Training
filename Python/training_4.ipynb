```python
#generator and iterators in python
1 iterator:
an iterator is an object that allows you to traverse through all 
the elements of a colletion (like list or tuple) one at a time
```

```python
key point:
implements __iter__() and __next__() methods
raises stopiteration when no more data is available
used in loops like for loop
```

```python
class My:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        if self.num<=5:
            x=self.num
            self.num+=1
            return x
        else:
            raise StopIteration
myc=My()
n=iter(myc)

for x in n:
    print(x)
```

```python
#generator
a generator is a special type of iterator that is defined using a function
and the 'yield' keyword
```

```python
keypoints

>uses yield instead of return
> generates values lazily
>more memory efficient than normal function
```

```python
def go():
    for i in range(1,6):
        yield i
gen=go()
for v in gen:
    print(v)
```

```python

from datetime import datetime
def go():
    for i in range(1,6):
        yield i
gen=go()
for v in gen:
    h=datetime.now()
    print(v)
    print(h)
```
