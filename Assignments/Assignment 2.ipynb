```python
2 where to use map, reduce, filter
```

```python
Answer: Map--- used when when we want to apply a function to every element
        reduce-when we want to reduce entire list into a single value
        filter-- keep only elements that satisfy a condition
```

```python
#map
l=[1,2,3,4,5,6]
s=list(map(lambda x:x**2 , l))
print(s)
```

```python
#reduce
from functools import reduce
l=[1,2,3,4,5,6]
s=reduce(lambda x,y:x+y,l)
print(s)
```

```python
#filter
l=[1,2,3,4,5,6]
s=list(filter(lambda x:x%2==0,l))
print(s)
```
